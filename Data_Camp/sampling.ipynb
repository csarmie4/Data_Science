#### Bias Any Stretch of the Imagination
Population: Complete dataset <br></br>
Sample: Subset of the dataset <br></br>
Collecting data by the easiest method is convenience sampling and can lead to bias sampling <br></br>
<b>np.random.beta(a=2, b=2, size=5000)</b> 5000 random numbers from 0 to 1(beta distribution) is created. Params named a and b <br></br> 
<b>np.random.normal(loc=2, scale=1.5, size=2)</b> loc and scale arguments set the mean and std of the distribution. <br></br>
<b>df.sample(n=10, random_state=21)</b> 5 random samples from df are selected. Argument random_state is similar to setting a random seed <br></br>

#### Don't get theory eyed
Systematic sampling: A sample value is picked every 5 samples \
<b>interval = var // var</b> Used below. Integer division, discards decimals \
<b>df.loc[::interval]</b> Systematic sampling. Double colon means nothing for the first argument, nothing for the second, and jump by interval. It gets every interval value of the sequence sliced. \
<b>shuffled= df.sample(frac=1)</b> Randomly shuffles the rows. Needed if doing systematic sampling and there is bias(patern) in the df. Doing this is the same as random sampling(.sample()) \
<b>shuffled = shuffled.reset_index(drop=True).reset_index()</b> drop=True clears the previous row indexes and chaining another reset_index creates new indexes. \ 
Filtering for a list of categories \
<b>var_0 = df["col1"].isin(list_var)</b> Used below. Subset before using var_1=df[var_0] \
<b>var_sample = var_0.sample(frac=0.1, random_state=212)</b> Takes a 10 percent simple random sample of data set \
<b>var_sample["col1"].value_counts(normalize=True)</b> To see the proportions of values in the sample \
Proportional stratified sampling: Makes sampling proportions similar to population proportions. \
To do this use groupby method before sampling the data set \
<b>var = df.groupby("col1").sample(frac=0.3, random_state=212)</b> \
Equal count stratified sampling \
Each category will have the same proportions. To do this switch frac=0.1 to n="int" \
<b>var = var_0.groupby("col1").sample(n=15, random_state=212)</b> Now each proportion is equal \
Weighted random sampling \
Can specify weights for any category to adjust relative probability \
<b>condition = var_0["col1"] == "cat_value"</b> Used below \
<b>var_0["new_col"] = np.where(condition, 2, 1)</b> Weight of 2 for rows that match condition and weight of 1 for the ones that do not. Now cat_value has a two time chance of being picked \
<b>var_0 = var_0.sample(frac=0.1, weights="weight")</b> Weights are now adjusted \
Cluster sampling: \
Use simple random sampling to pick soome subgroups then use simple random sampling on those subgroups \
<b>var_pop = list(df["col1"].unique)</b> Returns list of unique values \
<b>var_sample = random.sample(var_pop, k=3)</b> Randomly selectes 3 subgroups \
<b>condition = df["col1"].isin(var_sample)</b> Used below \ 
<b>df_cluster = df[condition]</b> Subset data set \
<b>df_cluster["col1"] = df_cluster["col1"].cat.remove_unused_categories()</b> Removes levels with zero rows \
Now we can do our sampling \
