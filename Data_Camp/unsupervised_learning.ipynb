#### Decorrelating your data and dimension reduction
Dimension reduction: More efficient storage and computation because it removes less-informative "noise" features <br></br>
PCA: Principal component analysis is a dimension reduction technique. <br></br>
Principal components = direction of variance
First step: PCA aligns data with axes and results in decorrelation(not lineraly correlated anymore) <br></br>
<b>from scipy.stats import pearsonr</b> Used below <br></br>
<b>pearsonr(x, y)</b> -1, 1 Used to measure correlation between x and y <br></br>

<b>from sklearn.decomposition import PCA</b> Used below <br></br>
Work flow: PCA(), .fit(), .transform() <br></br>
<b>range(model.n_components_)</b> Creates range enumerating the PCA features <br></br>
<b>features = pca.explained_variance_</b> Where the variance can be accessed of the PCA model <br></br>
<b>plt.bar(features, pca.explained_variance_)</b> To plt variance <br></br>
<b>PCA(n_components=2)</b> fit(), tansform(), Only keep 2 components/dimensions <br></br>
PCA assumes high variance info as informative and low variance as noise to be discarded. <br></br>
Not always true, for example word frequency arrays and are called "Sparse": most entries are zero <br></br>
Can use scipy.sparse.csr_matrix instead of numpy array. csr_matrix remembers only the non-zero entries which saves space <br></br>
PCA does not support csr_matrix have to use TruncatedSVD instead. They perform and can be interacted with the same way as PCA <br></br>




```python
# Perform the necessary imports
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Assign the 0th column of grains: width
width = grains[:, 0]

# Assign the 1st column of grains: length
length = grains[:, 1]

# Scatter plot width vs length
plt.scatter(width, length)
plt.axis('equal')
plt.show()

# Calculate the Pearson correlation
correlation, pvalue = pearsonr(width, length)

# Display the correlation
print(correlation)
```

```python
# Make a scatter plot of the untransformed points
plt.scatter(grains[:,0], grains[:,1])

# Create a PCA instance: model
model = PCA()

# Fit model to points
model.fit(grains)

# Get the mean of the grain samples: mean
mean = model.mean_

# Get the first principal component: first_pc
first_pc = model.components_[0, :]

# Plot first_pc as an arrow, starting at mean
plt.arrow(mean[0], mean[1], first_pc[0], first_pc[1], color='red', width=0.01)

# Keep axes on same scale
plt.axis('equal')
plt.show()
```

#### Discovering interpretable features
NMF: "non-negative matrix factorization" is a dimension reduction technique like PCA but they are interpretable(easier to explain) but samples have to be non-negative <br></br>
NMF must specify number of componentss and works with numpy arrays and csr_matrix <br></br>
<b>from sklearn.decomposition import NMF</b> Used below <br></br>
<b>NMF(n_components=2)</b> fit(), tansform(), Only keep 2 components/dimensions <br></br>


```python
# Import NMF
from sklearn.decomposition import NMF

# Create an NMF instance: model
model = NMF(n_components=6)

# Fit the model to articles
model.fit(articles)

# Transform the articles: nmf_features
nmf_features = model.transform(articles)

# Print the NMF features
print(nmf_features.round(2))
```
